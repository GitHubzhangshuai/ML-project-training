```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../data/data.csv')
df.head()
```

```python
df.shape
```

```python
df = df.drop('id',axis=1)
```

```python
df.diagnosis.unique()
```

```python
df['diagnosis']=df['diagnosis'].map({'M':1,'B':0})
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(20,20))
p = sns.heatmap(df.corr(),annot=True,square=True)
plt.show()
```

```python
print(df.diagnosis.value_counts())
p = df.diagnosis.value_counts().plot(kind='bar')
plt.show()
```

```python
x_data = df.drop(['diagnosis'],axis=1)
y_data = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_data,y_data,test_size=0.3,stratify=y_data)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,BaggingClassifier

classifiers = [
    KNeighborsClassifier(3),
    LogisticRegression(),
    MLPClassifier(hidden_layer_sizes=(20,50),max_iter=10000),
    DecisionTreeClassifier(),
    RandomForestClassifier(max_depth=9,min_samples_split=3),
    AdaBoostClassifier(),
    BaggingClassifier(),
]

log=[]
for clf in classifiers:
    clf.fit(x_train,y_train)
    name = clf.__class__.__name__
    print('='*30)
    print(name)
    print('*****Result****')
    test_predictions = clf.predict(x_test)
    acc = accuracy_score(y_test,test_predictions)
    print('Accuracy:{:.4%}' .format(acc))
    log.append([name,acc*100])
print('='*30)
```

```python
log = pd.DataFrame(log)
log
```

```python
log.rename(columns={0:'Classifiers',1:'Accuracy'},inplace=True)
```

```python
sns.barplot(x='Accuracy',y='classifiers',data=log,color='b')
plt.xlabel('Accuracy%')
plt.title('Classifier Accuracy')
plt.show()
```
