```python
import re
```

```python
text = open('../data/big.txt').read()
```

```python
text = re.findall('[a-z]+',text.lower())
```

```python
dic_words={}
for t in text:
    dic_words[t]=dic_words.get(t,0)+1
```

```python
dic_words
```

```python
alphabet='abcdefghijklmnopqrstuvwxyz'
def edit1(word):
    n=len(word)
    return set([word[0:i]+word[i+1:] for i in range(n)]+
              [word[0:i]+word[i+1]+word[i]+word[i+2:]for i in range(n-1)]+
              [word[0:i]+c+word[i+1:]for i in range(n) for c in alphabet]+
              [word[0:i]+c+word[i:]for i in range(n+1) for c in alphabet])
```

```python
def edit2(word):
    return set(e2 for e1 in edit1(word) for e2 in edit1(e1))
```

```python
e1 = edit1('something')
e2 = edit2('something')
len(e1)+len(e2)
```

```python
def known(words):
    w=set()
    for word in words:
        if word in dic_words:
            w.add(word)
    return w

def correct(word):
    candidates=known([word])or known(edit1(word))or known(edit2(word))or word
    if word==candidates:
        return word
    max_num=0
    for c in candidates:
        if dic_words[c]>=max_num:
            max_num=dic_words[c]
            candidate=c
    return candidate
```
